```python
import pandas as pd

from datetime import datetime

df = pd.read_csv('/content/drive/MyDrive/glassdoor_cleaned.csv')
```

```python
df.head()
```

```python
df['rating_date'] = pd.to_datetime(df['rating_date'], format='%Y-%m-%d')

df['rating_date'] = df['rating_date'].dt.strftime("%Y-%m-%dT%H:%M:%S.000Z")
df.head()
```

```python
print(df['rating_date'])
```

```python
df = df.drop(columns = ['summary'])
```

```python
columns_to_process = ['flags_ceo_approval', 'flags_business_outlook', 'flags_recommend_frend', 'employee_status']

for col in columns_to_process:

    df[col] = df[col].astype(str).str.upper()

df.head()
```

```python
df.to_csv('glassdoor_cleaned_V2.csv')
```
